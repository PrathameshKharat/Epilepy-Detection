## DATA PROFILING

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_train = pd.read_csv("epilepsy.txt")
```

```python
data_train.head()
```

```python
data= data_train.drop('name',axis =1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.shape
```

## FEATURE ENGINEERING

```python
data['MDVP:PPQ'] = data['MDVP:PPQ'].fillna(data['MDVP:PPQ'].median())
```

```python
data['Jitter:DDP'] = data['Jitter:DDP'].fillna(data['Jitter:DDP'].median())
```

```python
# FINDING CORRELATION BETWEEN DATA
data.corr()
```

```python
corrmat = data.corr()
# HEAT MAP
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
hm = sns.heatmap(data[top_corr_features].corr(),annot=True,cmap='Accent_r')
```

## MODELLING

```python
from sklearn.model_selection import train_test_split
```

```python
data.info()
```

```python
X = data.drop(['status'], axis = 1)
y = data['status']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier(random_state = 10)
random_forest_model.fit(X_train,y_train)
predict_train_data = random_forest_model.predict(X_train)

# print("Accurancy = (0:.3f)",format(metrics.accuracy_score(y_test,predict_train_data)))

```

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_accuracy = accuracy_score(predict_train_data,y_train)
```

```python
print("Training Data Accuracy:- ",training_data_accuracy*100)
```

```python
x_test = random_forest_model.predict(X_test)
```

```python
testing_data_accuracy = accuracy_score(x_test,y_test)
```

```python
print("Testing Data Accuracy: ",testing_data_accuracy*100)
```
